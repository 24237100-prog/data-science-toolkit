```python
import numpy as np

arr_224 = np.arange(16).reshape(2, 2, 4)
print("2×2×4 array:\n", arr_224)
```

```python
random_arr_34 = np.random.rand(3, 4)
print("3×4 random array:\n", random_arr_34)
```

```python
ones_arr = np.ones((4, 5))

zeros_arr = np.zeros((2, 3))
print("Array of ones:\n", ones_arr)
print("Array of zeros:\n", zeros_arr)
```

```python
arr_63 = np.random.rand(6, 3)

transposed_arr = arr_63.T
print("Original array shape:", arr_63.shape)
print("Transposed array shape:", transposed_arr.shape)
```

```python
n, p, m = 2, 3, 4

arr_np = np.random.rand(n, p)
arr_pm = np.random.rand(p, m)

dot_product = arr_np @ arr_pm
print("Dot product shape:", dot_product.shape)
print("Dot product result:\n", dot_product)
```
